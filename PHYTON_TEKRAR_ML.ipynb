#ABSTRACTION

Bir araba markası yeni bir araba modeli geliştirecektir.
Arabada bulunması gereken temel bileşenler (gaz pedalı, fren, gösterge paneli vb.) önceden bellidir.
Bu ortak yapı ve davranışları her araba türünde tekrar yazmak yerine abstract (ABC) bir üst sınıfta tanımlarız.
**Abstract class’tan nesne oluşturulamaz**; **`sadece alt sınıflar için bir şablon (sözleşme) görevi görür.`**
Elektrikli veya benzinli arabalar bu üst sınıftan türetilir ve zorunlu metotları kendi mantıklarına göre uygular.

```
from abc import ABC,abstractmethod #abstract kullanmak için gerekli olan kütüphanemiz

class Arac(ABC): #Abstract classtan türedi yani soyut oldu
  @abstractmethod
  def gitmek(self):# eğer başına @abstractmethod yazmazsak miras alan classlara zorunlu olmaz
    print("arac gider")
  @abstractmethod
  def istopEtmek(self):
    print("arac istop eder")

class lpgArac(Arac):
  def __init__(self):
    self.LPG=0
  def gitmek(self):
    return super().gitmek()
  def istopEtmek(self):
    print("ya da istersek başka bi şey döndeririz")
#lpgArac1=lpgArac() # nesne oluşturmaz çünkü abc methodları alt clasta oluşturmadın der
lpgArac1=lpgArac()
lpgArac1.gitmek()
lpgArac1.istopEtmek()
```
